```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
may_flights = flights.query("month == 'May'")
sns.lineplot(data=may_flights, x="year", y="passengers")
plt.savefig('01.png',dpi=500)
```

```python
flights_wide = flights.pivot("year", "month", "passengers")
flights_wide.head()
```

```python
sns.lineplot(data=flights_wide["May"])
plt.savefig('02.png',dpi=500)
```

```python
sns.lineplot(data=flights_wide)
plt.savefig('03.png',dpi=500)
```

```python
sns.lineplot(data=flights, x="year", y="passengers")
plt.savefig('04.png',dpi=500)
```

```python
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
plt.savefig('05.png',dpi=500)
```

```python
sns.lineplot(data=flights, x="year", y="passengers", hue="month", style="month")
plt.savefig('06.png',dpi=500)
```

```python
sns.lineplot(data=flights, x="passengers", y="year", orient="y")
plt.savefig('07.png',dpi=500)
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event")
plt.savefig('09.png',dpi=500)
```

```python
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="region", style="event")
plt.savefig('10.png',dpi=500)
```

```python
sns.lineplot(
    data=fmri,
    x="timepoint", y="signal", hue="event", style="event",
    markers=True, dashes=False
)
plt.savefig('11.png',dpi=500)
```

```python
sns.lineplot(
    data=fmri, x="timepoint", y="signal", hue="event", err_style="bars", errorbar=("se", 2),
)
plt.savefig('12.png',dpi=500)
```

```python
sns.lineplot(
    data=fmri.query("region == 'frontal'"),
    x="timepoint", y="signal", hue="event", units="subject",
    estimator=None, lw=1,
)
plt.savefig('13.png',dpi=500)
```

```python
dots = sns.load_dataset("dots").query("align == 'dots'")
dots.head()
```

```python
sns.lineplot(
    data=dots, x="time", y="firing_rate", hue="coherence", style="choice",
)
plt.savefig('14.png',dpi=500)
```

```python
sns.lineplot(
    data=dots.query("coherence > 0"),
    x="time", y="firing_rate", hue="coherence", style="choice",
     palette="flare"
)
plt.savefig('15.png',dpi=500)
```

```python
palette = sns.color_palette("mako_r", 6)
sns.lineplot(
    data=dots, x="time", y="firing_rate",
    hue="coherence", style="choice",
    palette=palette
)
plt.savefig('16.png',dpi=500)
```

```python
sns.lineplot(
    data=dots, x="time", y="firing_rate",
    size="coherence", hue="choice",
    legend="full"
)
plt.savefig('17.png',dpi=500)
```

```python
sns.lineplot(
    data=dots, x="time", y="firing_rate",
    size="coherence", hue="choice",
    sizes=(.25, 2.5)
)
plt.savefig('18.png',dpi=500)
```

```python
x, y = np.random.normal(size=(2, 5000)).cumsum(axis=1)
sns.lineplot(x=x, y=y, sort=False, lw=1)
plt.savefig('19.png',dpi=500)
```

```python
sns.relplot(
    data=fmri, x="timepoint", y="signal",
    col="region", hue="event", style="event",
    kind="line"
)
plt.savefig('20.png',dpi=500)
```
