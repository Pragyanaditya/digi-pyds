```python
import pandas as pd
import numpy as np
import seaborn as snb
import matplotlib.pyplot as plt
```

```python
iris=snb.load_dataset('iris')
iris.head()
```

```python
snb.histplot(iris, x="sepal_length",kde=True)
```

```python
snb.histplot(iris, x="sepal_width",kde=True)
```

```python
fig,ax=plt.subplots(figsize=(5,3))
snb.histplot(iris, x="sepal_width",kde=True,ax=ax)
plt.savefig('sepal_width_hist.png',dpi=300)
```

```python
snb.color_palette('rainbow')
```

```python
snb.set_palette('rainbow')
fig,ax=plt.subplots(figsize=(20,5),ncols=2,nrows=2)
snb.histplot(iris, x="sepal_width",kde=True,ax=ax[0,0],color='red')
snb.histplot(iris, x="sepal_length",kde=True,ax=ax[0,1])
snb.histplot(iris, x="petal_width",kde=True,ax=ax[1,0])
snb.histplot(iris, x="petal_length",kde=True,ax=ax[1,1])
```

```python
snb.barplot(data=iris,x='species',y='sepal_length')
```

```python
snb.barplot(iris,errorbar=None)
```

```python
snb.scatterplot(iris,x='sepal_length',y='sepal_width', hue='species')
```

```python
snb.jointplot(iris,x='sepal_length',y='sepal_width',hue='species')
```
